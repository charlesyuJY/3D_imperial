```python
def complete_keypoints(keypoints):
    if keypoints.shape[1] == 2:
        return np.hstack([
            keypoints, np.ones([keypoints.shape[0], 1]), np.zeros([keypoints.shape[0], 1])
        ])
    elif keypoints.shape[1] == 3:
        return np.hstack([
            keypoints, np.zeros([keypoints.shape[0], 1])
        ])
    else:
        return keypoints

def array_to_blob(array):
    return array.tostring()
```

```python

def insert_keypoints_descriptors (database_path,features_path):
  paths = types.SimpleNamespace()
  paths.database_path = os.path.join(database_path,'database.db')

  connection = sqlite3.connect(paths.database_path)
  cursor = connection.cursor()

  sum_num_features = 0
  cursor.execute('DELETE FROM keypoints;')
  cursor.execute('DELETE FROM descriptors;')
  images = {}
  cursor.execute('SELECT name, image_id FROM images;')

  for row in cursor:
      images[row[0]] = row[1]

  for image_name, image_id in images.items():
      keypoint_path = os.path.join(features_path, '%s.kpt.npy' % (image_name))
      descriptor_path = os.path.join(features_path, '%s.dsc.npy' % (image_name))
# load keypoints and descriptors
      keypoints = np.load(keypoint_path, allow_pickle=True)
      descriptors = np.load(descriptor_path, allow_pickle=True)
      descriptors = complete_keypoints(descriptors).astype(np.float32)

      if keypoints.shape[0] == 0:
          keypoints = np.zeros([0, 4])
      keypoints = complete_keypoints(keypoints).astype(np.float32)
      num_features = keypoints.shape[0]
      sum_num_features += num_features

# convert keypoints and dscriptors to binary format
      keypoints_str = array_to_blob(keypoints)
      descriptors_str = array_to_blob(descriptors)     
# load keypoints and descriptors back to the database
      cursor.execute(
          'INSERT INTO keypoints(image_id, rows, cols, data) VALUES(?, ?, ?, ?);',
          (image_id, keypoints.shape[0], keypoints.shape[1], keypoints_str)
      )

      cursor.execute(
        'INSERT INTO descriptors(image_id, rows, cols, data) VALUES(?, ?, ?, ?);',
        (image_id, descriptors.shape[0], descriptors.shape[1], descriptors_str)
      )

  connection.commit()
  cursor.close()
  connection.close()
```

```python
import torch


def mnn_similarity_matcher(descriptors1, descriptors2, threshold=0.8):
    # Mutual NN + similarity thresholding matcher for L2 normalized descriptors.
    device = descriptors1.device
    sim = descriptors1 @ descriptors2.t()

    # Retrieve first NN 1->2.
    match_sim, nn12 = torch.max(sim, dim=1)
    # Retrieve first NN 2->1.
    nn21 = torch.max(sim, dim=0)[1]
    
    # Mutual NN + similarity thresholding.
    ids1 = torch.arange(0, sim.shape[0], device=device)
    mask = torch.min(ids1 == nn21[nn12], match_sim >= threshold)

    # Final matches.
    matches = torch.stack([ids1[mask], nn12[mask]], dim=-1)
    match_sim = match_sim[mask]
    
    return (
        matches.data.cpu().numpy(),
        match_sim.data.cpu().numpy()
    )


def mnn_ratio_matcher(descriptors1, descriptors2, ratio=0.8):
    # Mutual NN + symmetric Lowe's ratio test matcher for L2 normalized descriptors.
    device = descriptors1.device
    sim = descriptors1 @ descriptors2.t()

    # Retrieve top 2 nearest neighbors 1->2.
    nns_sim, nns = torch.topk(sim, 2, dim=1)
    nns_dist = torch.sqrt(2 - 2 * nns_sim)
    # Compute Lowe's ratio.
    ratios12 = nns_dist[:, 0] / (nns_dist[:, 1] + 1e-8)
    # Save first NN and match similarity.
    nn12 = nns[:, 0]
    match_sim = nns_sim[:, 0]

    # Retrieve top 2 nearest neighbors 1->2.
    nns_sim, nns = torch.topk(sim.t(), 2, dim=1)
    nns_dist = torch.sqrt(2 - 2 * nns_sim)
    # Compute Lowe's ratio.
    ratios21 = nns_dist[:, 0] / (nns_dist[:, 1] + 1e-8)
    # Save first NN.
    nn21 = nns[:, 0]
    
    # Mutual NN + symmetric ratio test.
    ids1 = torch.arange(0, sim.shape[0], device=device)
    mask = torch.min(ids1 == nn21[nn12], torch.min(ratios12 <= ratio, ratios21[nn12] <= ratio))
    
    # Final matches.
    matches = torch.stack([ids1[mask], nn12[mask]], dim=-1)
    match_sim = match_sim[mask]

    return (
        matches.data.cpu().numpy(),
        match_sim.data.cpu().numpy()
    )
```

```python
def image_ids_to_pair_id(image_id1, image_id2):
    if image_id1 > image_id2:
        return 2147483647 * image_id2 + image_id1
    else:
        return 2147483647 * image_id1 + image_id2
```
