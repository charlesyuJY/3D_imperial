```python
# !pip3 install opencv-contrib-python
from platform import python_version

print(python_version())
import cv2
def extractImages(pathIn, pathOut):
    count = 0
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    success = True
    indent ='00'
    while success:
        if count>=10 and count<100:
          indent = '0'
        elif count>=100:
          indent = ''
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*500))    # added this line  time in ms/1000=number of frames
        success,image = vidcap.read()
        print ('Read a new frame: ', success)
        cv2.imwrite( pathOut + "/frame_%s%d_0.png" % (indent ,count), image)     # save frame as JPEG file
        count = count + 1
```

```python
pathIn = '/content/drive/MyDrive/3D_imperial_test_results/Video/V0690166_clipped.mp4'
pathOut= '/content/drive/MyDrive/3D_imperial_test_results/Output/EEE_first_floor/raw_images'
# pathIn = '/content/drive/MyDrive/3D_imperial_test_results/Video/Piccadilly.mp4'
# pathOut= '/content/drive/MyDrive/3D_imperial_test_results/Output/Picadilly_folder/Picadilly'
extractImages(pathIn,pathOut)
```
